```python
from pathlib import Path
from hydrometrics.stress_model_frac4 import StressConfig, run_stress

BASE_DIR = Path(".").resolve()
INPUT_DIR = BASE_DIR / "simulation_files"
TR = "1H"
OUT_DIR = BASE_DIR / "stress_results"
BESKRIVELSER_FILE = BASE_DIR / "/beskrivelser_final2.xlsx"

cfg = StressConfig(
    export_format="csv",

    # windows
    hours_before=48,
    hours_after=12,
    post_hyst_hours=72,

    # memory per group
    base_accum_mode="exp",
    base_tau_days=3.0,
    base_window_days=14,

    grad_accum_mode="exp",
    grad_tau_days=1,
    grad_window_days=14,

    forc_accum_mode="exp",
    forc_tau_days=1,
    forc_window_days=14,

    # base thresholds
    use_capacity_fraction=True,
    capacity_frac=0.5,

    use_quantile_threshold=False,
    gw_q_threshold=0.999,

    sm_use_quantile_threshold=False,
    sm_q_threshold=0.10,

    # OF threshold
    of_use_zero_threshold=True,
    of_use_quantile_threshold=False,
    of_q_threshold=0.90,

# --------------------------------------------------
# GRADIENT THRESHOLD MODE
# "quantile"        -> use pos_grad_q / neg_grad_q
# "fraction_of_max" -> use pos_grad_frac_of_max / neg_grad_frac_of_max
# --------------------------------------------------
    grad_threshold_mode="fraction_of_max",

# used when grad_threshold_mode="quantile"
    pos_grad_q=0.6,
    neg_grad_q=0.90,

# used when grad_threshold_mode="fraction_of_max"
# example: 0.7 means threshold at 70% of the max positive/negative gradient
    pos_grad_frac_of_max=0.15,
    neg_grad_frac_of_max=0.70,
    # forcing threshold choices
    # choose either:
    #   "fraction_of_max"
    #   "quantile"
    prec_threshold_mode="fraction_of_max",
    prec_frac_of_max=0.2,
    prec_q=0.80,

    swe_threshold_mode="fraction_of_max",
    swe_frac_of_max=0.10,
    swe_q=0.90,

    # toggles
    include_gw_base=True,
    include_sm_base=True,
    include_of_base=True,

    include_pos_grad=True,
    include_neg_grad=False,

    include_prec=True,
    include_swe=True,

    # internal pre-memory scalings
    gw_base_mult=1.0,
    sm_base_mult=1.0,
    of_base_mult=1.0,
    pos_grad_mult=1.0,
    neg_grad_mult=1.0,

    # windows used for accumulated forcing
    grad_hours=6,
    prec_hours=6,
    swe_hours=6,

    # final weighted-sum weights
    base_mult=1,
    grad_mult=1.0,
    prec_mult=1,
    swe_mult=1,
    of_mult=1,

    split_prec_by_temp=True,
    temp_rain_threshold_c=0.5,
    use_only_rain=True,

    combine_mode="sum",

    # normalization
    norm_q_low=0.0,
    norm_q_high=1.0,

    export_debug_intensities=True,
    export_debug_thresholds=True,

    # FINAL TOTAL STRESS COMBINATION
# "weighted_mean"           = your current behavior
# "weighted_sum_norm"       = weighted sum, then normalize final total to 0–1
# "weighted_sum_power_norm" = weighted sum of powered components, then normalize
    final_combine_mode="weighted_sum_norm",

# Only used if final_combine_mode="weighted_sum_power_norm"
# Larger values make peaks matter more and moderate values matter less
    final_combine_power=2.0,

)

run_stress(
    beskrivelser_file=BESKRIVELSER_FILE,
    input_dir=INPUT_DIR,
    tr=TR,
    out_dir=OUT_DIR,
    cfg=cfg,
)
```
